```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
```

```python
#NATIONAL: HOMICIDE BY YEAR
#change out the "demographic" variable
#can also do it by agency, region, or state

base_url = "https://api.usa.gov/crime/fbi/sapi/"
offense = "homicide"
demographic = "relationship" #count, ethnicity, race, sex, relationship
api_key = "?api_key=h6PvmzwIa7rXeB24Nrw1waxGedZEqqvOPdrU8MQb"

#Victim Data (national level): /api/data/nibrs/{offense}/victim/national/{variable}

getrequest = requests.get(base_url + "api/data/nibrs/" + offense + "/victim/national/" + demographic + api_key)
response = getrequest.json()
print(json.dumps(response, indent=4))
```

```python
#age to df example

data_year = []
age_0_9 = []
age_10_19 = []
age_20_29 = []
age_30_39 = []
age_40_49 = []
age_50_59 = []
age_60_69 = []
age_70_79 = []
age_80_89 = []
age_90_99 = []
unknown = []

for x in response:
    response = requests.get(base_url + "api/data/nibrs/" + offense + "/victim/national/" + demographic + api_key).json()
    data_year.append(response["results"][0]["data_year"])
    age_0_9.append(response["results"][0]["range_0_9"])
    age_10_19.append(response["results"][0]["range_10_19"])
    age_20_29.append(response["results"][0]["range_20_29"])
    age_30_39.append(response["results"][0]["range_30_39"])
    age_40_49.append(response["results"][0]["range_40_49"])
    age_50_59.append(response["results"][0]["range_50_59"])
    age_60_69.append(response["results"][0]["range_60_69"])
    age_70_79.append(response["results"][0]["range_70_79"])
    age_80_89.append(response["results"][0]["range_80_89"])
    age_90_99.append(response["results"][0]["range_90_99"])
    unknown.append(response["results"][0]["unknown"])

age_data = pd.DataFrame({"Year": data_year,
                         "0-9": age_0_9,
                         "10-19": age_10_19,
                         "20-29": age_20_29,
                         "30-39": age_30_39,
                         "40-49": age_40_49,
                         "50-59": age_50_59,
                         "60-69": age_60_69,
                         "70-79": age_70_79,
                         "80-89": age_80_89,
                         "90-99": age_90_99,
                         "Unknown": unknown})

age_data
```
